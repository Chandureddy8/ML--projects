# Hard Work Pay Off 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv(r"C:\Users\chandureddy\Downloads\Linear_X_Train.csv")
y = pd.read_csv(r"C:\Users\chandureddy\Downloads\Linear_y_Train.csv")
```

```python
#convert x,y to numpy arrays
x= x.values
y= y.values
```

```python
x
```

```python
#normalisation
u = x.mean()
std = x.std()
x = (x-u)/std
```

```python
#visualise
plt.style.use("seaborn")
plt.scatter(x,y)
plt.title("Hardwork vs performance graph")
plt.xlabel('Hardwork')
plt.ylabel("Performance")
plt.show()
```

```python
x.shape,y.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
#Return the coefficient of determination r**2 of the prediction.
reg.score(x,y)
```

```python
coeff = reg.coef_
coeff
```

```python
reg.intercept_
```

```python
plt.style.use("dark_background")
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red')
plt.show()
```
